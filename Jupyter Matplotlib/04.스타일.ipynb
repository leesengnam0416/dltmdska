# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1,2,3]
y = [2,4,8]

plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

# Marker

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o',linestyle='None')
```

```python
plt.plot(x,y, marker='v',markersize=10)
```

```python
plt.plot(x,y, marker='v', markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='v',markersize=15, markeredgecolor='blue',
        markerfacecolor='yellow')
```

# 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

# 색깔

```python
plt.plot(x,y, color='b')
```

# 포맷

```python
plt.plot(x,y, 'ro--') # color , marker , linestyle
```

```python
plt.plot(x,y, 'bv:')
```

# 축약어

```python
plt.plot(x,y, marker='o' , mfc='red', ms=10,mec='blue',ls=':')
```

# 투명도

```python
plt.plot(x,y, marker='o',mfc='red',ms=10, alpha=0.4)
```

# 그래프의 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=150)
plt.plot(x,y)
```

# 배경색

```python
plt.figure(facecolor='r') # red 도 가능
plt.plot(x,y)
```
