## 9.데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df=  pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt]
```

```python
df[df['키'] >= 185]
```

```python
df.loc[df['키'] >= 185 , '수학']
```

```python
df.loc[df['국어'] >= 50 , ['이름','과학','수학']]
```

## 다양한 조건

```python
df.loc[ (df['키'] >=185) & (df['학교'] == '북산고')]
```

```python
df
```

```python
df.loc[(df['학교'] == '북산고') & (df['수학'] <= 50)]
```

```python
df.loc[(df['키'] >= 185) | (df['수학'] <= 50)]
```

## Str 함수

```python
filt= df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
df[~filt]
```

```python
langs = ['Python',"Java"]
filt = df["SW특기"].isin(langs)
df[filt]
```

```python
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt]
```
