## 12.데이터 수정

```python
import pandas as pd
df=  pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

## column 수정

```python
df['학교'].replace({'북산고':'상북고','능남고':'무슨고'})
```

```python
df['학교'].replace({'북산고':'상북고'}, inplace=True)
```

```python
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['학교'] = df['학교'] + '등학교'
```

```python
df
```

## column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['결과'] = 'Fail'
df
```

```python
df.loc[df['총합'] > 400 , '결과'] = 'Pass'
df
```
